```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from sklearn.linear_model import LinearRegression
```

```python
data=xr.open_dataset('g4.areaAvgTimeSeries.MODISA_L3m_SST4_Monthly_9km_R2019_0_sst4.20020101-20231231.2E_3N_8E_6N.nc')
data
```

```python

x=data.datamonth
y=data.MODISA_L3m_SST4_Monthly_9km_R2019_0_sst4
x=np.array(x).reshape(-1,1)
y=np.array(y)
model=LinearRegression().fit(x,y)
print(f'intercept:{model.intercept_}')

y_fit=model.intercept_ + model.coef_ *x
plt.plot(data.time,data.MODISA_L3m_SST4_Monthly_9km_R2019_0_sst4,color="red")
plt.plot(data.time,y_fit,label='Trend Line',color= "black")
```
